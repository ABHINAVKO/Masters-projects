```r
# Graph 4

# Tutotorial on plotting US map in R : http://eriqande.github.io/rep-res-web/lectures/making-maps-with-R.html

library('ggplot2')
library('ggmap')
library('maps')
library('mapdata')
library('stringr')

states <- map_data("state")

data <- read.csv("HeatMap 1 Data.csv")
levels(data$STATENAME) <- tolower(levels(data$STATENAME))
data$ACTIVITY.LEVEL <- str_replace_all(data$ACTIVITY.LEVEL, 'Level ', '')

d <- merge(x = states, y = data, by.x = "region", by.y = "STATENAME", all.x = TRUE)

ggplot(data = d) + 
  geom_polygon(aes(x = long, y = lat, fill = ACTIVITY.LEVEL, group = group), color = "white") + 
  coord_fixed(1.3) +
  guides(fill=FALSE)
```
